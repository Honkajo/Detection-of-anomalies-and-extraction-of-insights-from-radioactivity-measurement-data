The preprocessing dataset shows substantial variability across samples, indicating that multiple preprocessing workflows are being applied rather than a single uniform procedure. Several key variables, such as weight, mass, and quantum, are highly skewed with extreme maximum values, suggesting the need for transformation before clustering. A few values, particularly in density and dryWeightPercent, appear physically unrealistic and may represent data quality issues that require further review. Many advanced preprocessing variables such as ashing and sieving are present only for subsets of samples, confirming workflow heterogeneity. Overall, the dataset contains meaningful structural differences that are likely to produce distinct natural clusters even without using sample class information.

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = d1_ready.select_dtypes(include=["number"])

corr = numeric_df.corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Correlation Matrix")
plt.show()
```

The correlation matrix shows strong redundancy among mass related variables. In particular, massInJar, massInJarGross, and dryWeightNet are highly correlated, indicating they capture similar information about sample size and should not all be used simultaneously in clustering. DryWeightPercent and moisture_ratio are strongly negatively correlated, which is expected since moisture_ratio is derived from dryWeightPercent. AshWeightNet and ash_to_dry also show high correlation, meaning one of them is sufficient to represent mineral content. Sieved variables correlate strongly among themselves but are present only for a small subset of samples, indicating specialized preprocessing workflows. Overall, the matrix confirms that feature reduction is necessary before clustering to avoid redundancy and dominance of mass related variables.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define representative preprocessing features
# These were selected after correlation analysis to avoid redundancy
key_vars = [
    "weightnet",         # sample size
    "height",            # packing / jar structure
    "density",           # physical composition
    "moisture_ratio",    # water content proxy
    "ash_to_dry",        # mineral fraction proxy
    "jar_check_error"    # mass balance consistency check
]

# Loop through each selected feature
for col in key_vars:
    
    # Only proceed if column exists in the dataset
    if col in d1_ready.columns:
        
        plt.figure()  # create new plot
        
        # Replace infinite values with NaN and remove missing values
        # This prevents plotting errors from division-based features
        data = d1_ready[col].replace([np.inf, -np.inf], np.nan).dropna()
        
        # If the variable is strictly positive,
        # apply log transformation to reduce skewness
        if data.min() > 0:
            data = np.log1p(data)
        
        # Plot histogram to inspect distribution shape
        data.hist(bins=50)
        
        # Title shows whether log scaling may have been applied
        plt.title(col + " (log scaled if positive)")
        
        plt.show()
```

```python
# -----------------------------
# 1) Detect negative mass values
# -----------------------------

# List of physical mass-related columns
mass_cols = [
    "weightnet",        # original sample mass
    "massinjar",        # sample mass inside jar
    "massinjartare",    # empty jar weight
    "massinjargross",   # jar + sample
    "dryweightnet",     # mass after drying
    "ashweightnet"      # mass after ashing
]

# Create a rule flag for each column: True if value is negative
for col in mass_cols:
    if col in d1_ready.columns:
        d1_ready[f"rule_negative_{col}"] = d1_ready[col] < 0
```

```python
# ---------------------------------------
# 2) Dry weight should not exceed original
# ---------------------------------------

if "dryweightnet" in d1_ready.columns and "weightnet" in d1_ready.columns:
    
    # Flag rows where drying increased mass (physically suspicious)
    d1_ready["rule_dry_gt_original"] = (
        d1_ready["dryweightnet"] > d1_ready["weightnet"]
    )
```

```python
# ---------------------------------------
# 3) Percent must be between 0 and 100
# ---------------------------------------

if "dryweightpercent" in d1_ready.columns:
    
    # Flag rows where percentage is outside valid range
    d1_ready["rule_invalid_percent"] = (
        (d1_ready["dryweightpercent"] < 0) |
        (d1_ready["dryweightpercent"] > 100)
    )
```

```python
# ---------------------------------------
# 4) Mass balance consistency check
# ---------------------------------------

if all(col in d1_ready.columns for col in ["massinjargross", "massinjartare", "massinjar"]):
    
    tolerance = 0.01  # small allowed measurement error
    
    # Expected gross mass = tare + net
    expected = d1_ready["massinjartare"] + d1_ready["massinjar"]
    
    # Flag rows where actual gross deviates beyond tolerance
    d1_ready["rule_mass_balance_error"] = (
        (d1_ready["massinjargross"] - expected).abs() > tolerance
    )
```

```python
# ---------------------------------------
# 5) Density plausibility check
# ---------------------------------------

if "density" in d1_ready.columns:
    
    # Density must be positive and within realistic upper bound
    d1_ready["rule_density_unrealistic"] = (
        (d1_ready["density"] <= 0) |
        (d1_ready["density"] > 20)
    )
```

```python
# ---------------------------------------
# 6) Combine all rule flags
# ---------------------------------------

# Collect all columns that start with "rule_"
rule_cols = [c for c in d1_ready.columns if c.startswith("rule_")]

# Mark row as probably erroneous if ANY rule is True
d1_ready["probably_erroneous"] = d1_ready[rule_cols].any(axis=1)
```

```python
d1_clean = d1_ready[~d1_ready["probably_erroneous"]].copy()

print("Original rows:", len(d1_ready))
print("Clean rows:", len(d1_clean))
```

```python
# Final feature set selected after:
# - Correlation analysis
# - Distribution inspection
# - Removing redundancy
# - Adding derived features

clustering_features = [
    "weightnet",        # sample size
    "height",           # jar packing structure
    "density",          # physical composition
    "moisture_ratio",   # water content
    "ash_to_dry",       # mineral fraction
    "jar_check_error"   # mass consistency check
]

# Keep only features that actually exist
clustering_features = [
    f for f in clustering_features if f in d1_clean.columns
]
```

```python
# Extract selected features
X = d1_clean[clustering_features].copy()

# Replace infinite values caused by divisions
X = X.replace([np.inf, -np.inf], np.nan)
```

```python
from sklearn.impute import SimpleImputer

# Use median imputation to avoid distortion from outliers
imputer = SimpleImputer(strategy="median")

X_imputed = imputer.fit_transform(X)
```

```python
import numpy as np

# Apply log1p only to features that are strictly positive
# This reduces skewness and stabilizes clustering
for i, col in enumerate(clustering_features):
    if X_imputed[:, i].min() > 0:
        X_imputed[:, i] = np.log1p(X_imputed[:, i])
```

```python
from sklearn.preprocessing import StandardScaler

# Scale features so each has mean 0 and variance 1
# Prevents large-scale variables from dominating clustering
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X_imputed)
```

```python
from sklearn.cluster import KMeans

# Initialize KMeans with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)

# Fit and predict cluster labels
clusters = kmeans.fit_predict(X_scaled)

# Save cluster labels back to dataset
d1_clean["cluster"] = clusters

# Check cluster sizes
d1_clean["cluster"].value_counts()
```

```python
import numpy as np

# moisture_ratio = 1 - (dryWeightPercent / 100)
if "dryweightpercent" in d1_clean.columns:
    d1_clean["moisture_ratio"] = 1 - (d1_clean["dryweightpercent"] / 100)

# ash_to_dry = ashWeightNet / dryWeightNet
if "ashweightnet" in d1_clean.columns and "dryweightnet" in d1_clean.columns:
    d1_clean["ash_to_dry"] = d1_clean["ashweightnet"] / d1_clean["dryweightnet"]

# jar_check_error = massInJarGross - (massInJarTare + massInJar)
if all(c in d1_clean.columns for c in ["massinjargross", "massinjartare", "massinjar"]):
    d1_clean["jar_check_error"] = d1_clean["massinjargross"] - (
        d1_clean["massinjartare"] + d1_clean["massinjar"]
    )

# Replace inf values created by division
d1_clean = d1_clean.replace([np.inf, -np.inf], np.nan)
```

```python
import numpy as np
import pandas as pd

cols = ["weightnet", "height", "density", "moisture_ratio", "ash_to_dry", "jar_check_error"]
cols = [c for c in cols if c in d1_clean.columns]

# remove weird spacing and commas if any, then convert
for c in cols:
    d1_clean[c] = (
        d1_clean[c]
        .astype(str)
        .str.replace(",", ".", regex=False)
        .str.strip()
    )
    d1_clean[c] = pd.to_numeric(d1_clean[c], errors="coerce")

# replace inf just in case
d1_clean[cols] = d1_clean[cols].replace([np.inf, -np.inf], np.nan)

# now mean will work
d1_clean.groupby("cluster")[cols].mean()
```

```python
cols = ["weightnet", "height", "density", "moisture_ratio", "ash_to_dry", "jar_check_error"]
cols = [c for c in cols if c in d1_clean.columns]

cluster_profile = d1_clean.groupby("cluster")[cols].agg(["count", "mean", "median"])
cluster_profile
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
import numpy as np

X = d1_clean[cols].copy()
X = X.replace([np.inf, -np.inf], np.nan)

imputer = SimpleImputer(strategy="median")
X_imp = imputer.fit_transform(X)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imp)

iso = IsolationForest(
    n_estimators=300,
    contamination=0.02,
    random_state=42
)

scores = -iso.fit_predict(X_scaled)
d1_clean["anomaly_label"] = iso.predict(X_scaled)      # -1 anomaly, 1 normal
d1_clean["anomaly_score"] = -iso.decision_function(X_scaled)  # higher = more abnormal
```

```python
def classify_row(r):
    if r.get("probably_erroneous", False):
        return "Probably erroneous"
    if r.get("anomaly_label", 1) == -1:
        return "Atypical but plausible"
    return "Typical"

d1_clean["preprocess_class"] = d1_clean.apply(classify_row, axis=1)

d1_clean["preprocess_class"].value_counts()
```
